```python
names =["Kiran", "fatima", "aqsa", "moiz", "zahra","hamza","manahil","uzma"]
departments =["javascript","python","html","css","ai engineer","data science","HR", "Engineering", "Sales", "Marketing", "Finance"]
skills =[["html","css","JavaScript"],["python","JavaScript","ai engineer"],["data science","next js"]]

employees = []
import random
for i in range(1,10001):
    emp = {
        "id": i,
        "name": names[i % len(names)],
        "age": random.randint(20,60),
        "department": departments[i % len(departments)],
        "skill" : skills[i% len(skills)],
        "salary": random.randint(20000,600000)
    }
    employees.append(emp)
for e in employees[:3]:
    print(e)
```

```python
# call the data
print(employees[2]["salary"])

for j in range(10,20):
    print(employees[j]["age"])

print(employees[4]["name"])
```

```python
#add the data
print(employees[4]["skill"].append("js"))
for e in employees[:6]:
    print(e)
```

```python
# update the data
employees[2]["salary"]=10009
for e in employees[:3]:
    print(e)
```

```python
# other things
# first 10 employees names

for i in employees[:10]:
    print(i["name"])
```

```python

for emp in employees:
    if emp["age"] < 25:
        print("Name:", emp["name"], "- Age:", emp["age"])
```

```python
# count how many employees are age of 30
count=0
for emp in employees:
    if emp["age"] == 30:
        count+=1
print(count)
```

```python
#employee names who work in the "Sales" department:
for emp in employees:
    if emp["department"] == "Sales":
        print(emp["name"])
```
